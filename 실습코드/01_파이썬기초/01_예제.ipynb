```python
age = 25            # 정수 int
height = 173.5      # 실수 float
name = '홍길동'      # 문자열 str
is_student = True   # 불리언 bool (True/False)
print(age, height, name, is_student)
print(type(age), type(height), type(name), type(is_student))
```

```python
score = 92
print(f'{name}님의 점수는 {score}점입니다.')
print(f'평균과의 차이: {score - 80}점')
```

```python
scores = [90, 85, 77, 60, 100]
print('첫 번째:', scores[0])     # 0부터 시작
print('마지막:', scores[-1])     # 음수는 뒤에서부터
print('앞 3개:', scores[0:3])    # 슬라이싱 (끝 제외)
print('개수:', len(scores))
scores.append(88)               # 값 추가
print('추가 후:', scores)
```

```python
person = {'이름': '홍길동', '나이': 25, '점수': 92}
print(person['이름'])          # 키로 값 꺼내기
print(person.keys())           # 키 목록
print(person.values())         # 값 목록
person['합격'] = True           # 새 키 추가
print(person)
```

```python
score = 82
if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
else:
    grade = 'C'
print(f'{score}점 → {grade}')
```

```python
scores = [90, 85, 77]
for s in scores:                 # 리스트 순회
    print('점수:', s)
total = 0
for s in scores:
    total = total + s
print('합계:', total, '/ 평균:', total / len(scores))
for i in range(3):               # 0,1,2
    print('반복', i)
```

```python
def get_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    return 'C'
print(get_grade(95))
print(get_grade(83))
print(get_grade(70))
```

```python
scores = [90, 85, 77, 60]
passed = [s for s in scores if s >= 80]   # 80 이상만
doubled = [s * 2 for s in scores]         # 각 값 2배
print('합격:', passed)
print('2배:', doubled)
```

```python
# 1) 연산자 -- 조건문/필터의 재료
print(7 // 3, 7 % 3, 2 ** 3)              # 몫, 나머지, 거듭제곱
print(3 == 3, 3 != 5, 3 < 5, 5 >= 5)      # 비교 → True/False
print(True and False, True or False, not True)   # 논리 and/or/not
print(90 in [90, 85], 'age' in {'age': 25})      # 멤버십 in
```

```python
# 2) while -- 조건이 참인 동안 반복
n = 3
while n > 0:
    print(n)
    n = n - 1        # 조건을 바꿔야 멈춘다 (안 그러면 무한 반복)
```

```python
# 3) 튜플 -- 값 묶음(수정 불가). 함수 다중 반환이 이 튜플
point = (3, 5)
x, y = point               # 언패킹
print(x, y)
```

```python
# 4) 중첩 자료구조 -- 리스트 안 딕셔너리 = DataFrame의 재료
rows = [{'이름': '홍길동', '점수': 90}, {'이름': '김철수', '점수': 85}]
print(rows[0]['이름'])      # 홍길동
```

```python
# 5) 딕셔너리 안전 접근 & None
d = {'age': 25}
print(d.get('city'))              # 없는 키 → None (에러 안 남)
print(d.get('city', '서울'))       # 없으면 기본값 '서울'
print(None is None, bool(0), bool(''))   # None 확인 / 0·빈문자 = False
```
