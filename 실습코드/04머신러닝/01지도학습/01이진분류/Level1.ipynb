```python
# ############################################################
# Level 1 — 유방암 양성/악성 분류 (가장 단순)
# ############################################################
# ------------------------------------------------------------
# [목적] 도구 불러오기
# ------------------------------------------------------------
from sklearn.datasets import load_breast_cancer   # 유방암 데이터 (연습용, 0=악성/1=양성)
from sklearn.model_selection import train_test_split   # 학습/시험 나누기
from sklearn.linear_model import LogisticRegression    # 로지스틱 회귀 (확률로 분류하는 기본 모델)
from sklearn.metrics import accuracy_score         # 정확도 (맞힌 비율)
```

```python
# ------------------------------------------------------------
# [목적] 데이터 준비 + 학습/시험 8:2 분리
# ------------------------------------------------------------
X, y = load_breast_cancer(return_X_y=True)         # 특성 X, 정답 y (0=악성, 1=양성)
X_tr, X_te, y_tr, y_te = train_test_split(
    X, y, test_size=0.2, random_state=0, stratify=y)   # 비율 유지하며 8:2로 (결과 고정)
```

```python
# ------------------------------------------------------------
# [목적] 학습 후 정확도 확인
# ------------------------------------------------------------
model = LogisticRegression(max_iter=5000).fit(X_tr, y_tr)   # 학습 (max_iter=반복 넉넉히)
print('정확도:', round(accuracy_score(y_te, model.predict(X_te)), 3))  # 시험셋 정확도
```
