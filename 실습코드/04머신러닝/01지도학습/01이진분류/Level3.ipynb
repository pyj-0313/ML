```python
# ############################################################
# Level 3 — 타이타닉 생존 예측 (공개 데이터, 글자 -> 숫자)
# ############################################################
# ------------------------------------------------------------
# [목적] 도구 불러오기 — 표 / 분리 / 모델 / 평가
# ------------------------------------------------------------
import pandas as pd                               # 표(CSV) 다루기
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier   # 랜덤포레스트 (나무들의 다수결)
from sklearn.metrics import accuracy_score, roc_auc_score
```

```python
# ------------------------------------------------------------
# [목적] 데이터 불러오기 + 전처리 (결측 채우기 / 글자 -> 0·1)
# ------------------------------------------------------------
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)                              # 공개 URL에서 바로 읽기 (다운로드 불필요)
df['Age'] = df['Age'].fillna(df['Age'].median())  # 나이 결측 채우기 (빈칸을 중앙값으로)
X = pd.get_dummies(df[['Pclass','Sex','Age','SibSp','Parch','Fare']])  # 글자(성별 등) -> 0/1 열
y = df['Survived']                                # 정답 (1=생존, 0=사망)
```

```python
# ------------------------------------------------------------
# [목적] 8:2 분리 -> 학습 -> 정확도 + ROC-AUC
# ------------------------------------------------------------
X_tr, X_te, y_tr, y_te = train_test_split(
    X, y, test_size=0.2, random_state=0, stratify=y)              # 비율 유지 8:2
model = RandomForestClassifier(n_estimators=200, random_state=0).fit(X_tr, y_tr)  # 나무 200그루 학습
print('정확도 :', round(accuracy_score(y_te, model.predict(X_te)), 3))            # 맞힌 비율
print('ROC-AUC:', round(roc_auc_score(y_te, model.predict_proba(X_te)[:, 1]), 3)) # 확률 순위 품질
```

```python
# ============================================================
# [결과 해석]
#  · 정확도 ~ 0.79, ROC-AUC ~ 0.82 -> 생존 여부를 꽤 잘 맞힘
#  · 글자 열(성별 등)은 get_dummies로 0/1 바꿔야 학습 가능, 나이 결측은 미리 채움
#  · 실제 캐글 데이터라 내장 데이터(Level1·2)보다 점수가 낮음 (현실 데이터의 노이즈)
# ============================================================
```
