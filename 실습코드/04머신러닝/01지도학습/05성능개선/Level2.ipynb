```python
# ############################################################
# Level 2 — 교차검증 + GridSearchCV 튜닝
# ############################################################
# ------------------------------------------------------------
# [목적] 도구 불러오기
# ------------------------------------------------------------
from sklearn.datasets import load_diabetes
from sklearn.model_selection import cross_val_score, GridSearchCV   # 교차검증 / 자동 튜닝
from sklearn.linear_model import Ridge            # 정규화 회귀
```

```python
# ------------------------------------------------------------
# [목적] 교차검증 — 5번 나눠 평균 (믿을 수 있는 점수)
# ------------------------------------------------------------
X, y = load_diabetes(return_X_y=True)
cv = cross_val_score(Ridge(alpha=1), X, y, cv=5, scoring='r2')   # 5겹 교차검증
print('교차검증 평균 R2:', round(cv.mean(), 3))   # 평균 성적
```

```python
# ------------------------------------------------------------
# [목적] 튜닝 — alpha 후보를 자동으로 모두 시도해 최고 선택
# ------------------------------------------------------------
grid = GridSearchCV(Ridge(), {'alpha': [0.01, 0.1, 1, 10, 100]}, cv=5, scoring='r2')
grid.fit(X, y)                                    # 후보 5개를 교차검증으로 비교
print('최적 alpha:', grid.best_params_, '| 최적 R2:', round(grid.best_score_, 3))
```
