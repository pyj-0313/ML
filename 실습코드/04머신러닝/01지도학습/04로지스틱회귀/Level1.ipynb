```python
# ############################################################
# Level 1 — 유방암 양성/악성 분류 (로지스틱 기본)
# ############################################################
# ------------------------------------------------------------
# [목적] 도구 불러오기
# ------------------------------------------------------------
from sklearn.datasets import load_breast_cancer   # 유방암 데이터 (0=악성/1=양성)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score        # 정확도
```

```python
# ------------------------------------------------------------
# [목적] 데이터 준비 + 8:2 분리
# ------------------------------------------------------------
X, y = load_breast_cancer(return_X_y=True)        # 특성 X, 정답 y(0/1)
X_tr, X_te, y_tr, y_te = train_test_split(
    X, y, test_size=0.2, random_state=0, stratify=y)   # 비율 유지 8:2
```

```python
# ------------------------------------------------------------
# [목적] 학습 후 정확도 확인
# ------------------------------------------------------------
model = LogisticRegression(max_iter=5000).fit(X_tr, y_tr)   # 학습 (반복 넉넉히)
print('정확도:', round(accuracy_score(y_te, model.predict(X_te)), 3))
```

```python
# ============================================================
# [결과 해석]
#  · 정확도 ~ 0.947 -> 양성/악성을 약 95% 맞힘
#  · 로지스틱은 확률을 내부에서 계산해 0.5 기준으로 0/1 판정
#  · 다음 단계(Level 2)에서 '계수'로 판단 근거까지 들여다봄
# ============================================================
```
