```python
# ############################################################
# Level 2 — 계수로 '이유' 읽기 (어떤 특성이 판단을 가르나)
# ############################################################
# ------------------------------------------------------------
# [목적] 도구 불러오기 — 표 / 데이터 / 분리 / 전처리 / 모델
# ------------------------------------------------------------
import pandas as pd                               # 계수를 표로 정렬해 보기
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler  # 스케일링 (계수 비교를 공정하게)
from sklearn.linear_model import LogisticRegression
```

```python
# ------------------------------------------------------------
# [목적] 데이터 준비 + 분리 + 스케일링(계수 크기 비교 위해 필수)
# ------------------------------------------------------------
data = load_breast_cancer()
X_tr, X_te, y_tr, y_te = train_test_split(
    data.data, data.target, test_size=0.2, random_state=0, stratify=data.target)
sc = StandardScaler()
X_tr_s = sc.fit_transform(X_tr)                   # 같은 잣대로 (계수 크기를 공정하게 비교)
```

```python
# ------------------------------------------------------------
# [목적] 학습 후 계수 정렬 — 부호=방향, 크기=영향력
# ------------------------------------------------------------
model = LogisticRegression(max_iter=5000).fit(X_tr_s, y_tr)   # 학습
coef = pd.Series(model.coef_[0], index=data.feature_names).sort_values()  # 특성별 계수 정렬

print('계수 낮은(작을수록 악성 쪽) 3개:'); print(coef.head(3).round(2))   # 음수 큰 특성
print('계수 높은(클수록 양성 쪽) 3개:');  print(coef.tail(3).round(2))   # 양수 큰 특성
```
