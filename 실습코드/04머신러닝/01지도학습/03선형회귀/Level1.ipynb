```python
# ############################################################
# Level 1 — 당뇨 진행도 예측 (가장 단순, R2)
# ############################################################
# ------------------------------------------------------------
# [목적] 도구 불러오기
# ------------------------------------------------------------
from sklearn.datasets import load_diabetes        # 당뇨 데이터 (1년 뒤 진행도=숫자)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression # 선형회귀
from sklearn.metrics import r2_score              # R2 (설명력)
```

```python
# ------------------------------------------------------------
# [목적] 데이터 준비 + 8:2 분리
# ------------------------------------------------------------
X, y = load_diabetes(return_X_y=True)             # 특성 X, 정답 y(숫자)
X_tr, X_te, y_tr, y_te = train_test_split(
    X, y, test_size=0.2, random_state=0)          # 8:2 (회귀라 stratify 없음)
```

```python
# ------------------------------------------------------------
# [목적] 학습 후 R2 확인
# ------------------------------------------------------------
model = LinearRegression().fit(X_tr, y_tr)        # 학습(직선 찾기)
print('R2:', round(r2_score(y_te, model.predict(X_te)), 3))   # 1에 가까울수록 잘 설명
```

```python
# ============================================================
# [결과 해석]
#  · R2 ~ 0.332 -> 당뇨 진행도 변동의 약 33%만 설명 (낮은 편)
#  · 주어진 특성만으로는 예측이 쉽지 않은 문제라는 뜻
#  · R2가 낮다고 코드가 틀린 건 아님 -> 데이터·특성의 한계
# ============================================================
```
