```python
# ############################################################
# Level 2 — 캘리포니아 집값 예측 + MAE / RMSE / R2
# ############################################################
# ------------------------------------------------------------
# [목적] 도구 불러오기 — 숫자 / 데이터 / 분리 / 모델 / 평가 3종
# ------------------------------------------------------------
import numpy as np                                # 제곱근 계산
from sklearn.datasets import fetch_california_housing   # 캘리포니아 집값(처음엔 다운로드)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# ------------------------------------------------------------
# [목적] 데이터 준비 + 8:2 분리
# ------------------------------------------------------------
X, y = fetch_california_housing(return_X_y=True)  # 특성 8개, 정답=집값(10만 달러 단위)
X_tr, X_te, y_tr, y_te = train_test_split(
    X, y, test_size=0.2, random_state=0)          # 8:2 분할
```

```python
# ------------------------------------------------------------
# [목적] 학습 -> 예측 -> 세 지표 함께 보기
# ------------------------------------------------------------
model = LinearRegression().fit(X_tr, y_tr)        # 학습
pred = model.predict(X_te)                        # 예측
print('MAE :', round(mean_absolute_error(y_te, pred), 3))          # 평균 오차
print('RMSE:', round(np.sqrt(mean_squared_error(y_te, pred)), 3))  # 큰 오차 강조
print('R2  :', round(r2_score(y_te, pred), 3))                     # 설명력
```

```python
# ============================================================
# [결과 해석]
#  · MAE ~ 0.535 -> 평균 약 0.53 (= 집값 단위 10만 달러 기준 약 5.3만 달러) 빗나감
#  · RMSE ~ 0.727 (> MAE) -> 크게 틀린 지역이 일부 있음
#  · R2 ~ 0.594 -> 집값 변동의 약 59%를 설명 (특성 8개치고 무난)
# ============================================================
```
