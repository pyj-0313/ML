```python
# ############################################################
# Level 1 — 붓꽃 3품종 분류 (가장 단순)
# ############################################################
# ------------------------------------------------------------
# [목적] 도구 불러오기
# ------------------------------------------------------------
from sklearn.datasets import load_iris            # 붓꽃 데이터 (3품종)
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier    # 랜덤포레스트
from sklearn.metrics import accuracy_score         # 정확도
```

```python
# ------------------------------------------------------------
# [목적] 데이터 준비 + 8:2 분리
# ------------------------------------------------------------
X, y = load_iris(return_X_y=True)                 # 특성 X, 정답 y (품종 0/1/2)
X_tr, X_te, y_tr, y_te = train_test_split(
    X, y, test_size=0.2, random_state=0, stratify=y)   # 품종 비율 유지 8:2
```

```python
# ------------------------------------------------------------
# [목적] 학습 후 정확도 확인
# ------------------------------------------------------------
model = RandomForestClassifier(random_state=0).fit(X_tr, y_tr)   # 학습
print('정확도:', round(accuracy_score(y_te, model.predict(X_te)), 3))  # 시험셋 정확도
```

```python
# ============================================================
# [결과 해석]
#  · 정확도 ~ 0.933 -> 세 품종을 약 93% 맞힘
#  · 세 품종 개수가 균형이라 macro와 weighted가 거의 같음
#  · 다중분류도 이진분류와 코드는 거의 동일 (모델이 알아서 3개로 나눔)
# ============================================================
```
