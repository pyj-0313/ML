```python
# ############################################################
# Level 3 — 펭귄 종 분류 (공개 데이터, 정답이 글자)
# ############################################################
# ------------------------------------------------------------
# [목적] 도구 불러오기 — 표 / 분리 / 모델 / 평가
# ------------------------------------------------------------
import pandas as pd                               # 표(CSV) 다루기
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
# ------------------------------------------------------------
# [목적] 데이터 불러오기 + 정리 (결측 행 제거)
# ------------------------------------------------------------
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv'
df = pd.read_csv(url).dropna()                    # 공개 URL에서 읽고 빈칸 있는 행 제거
X = df[['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']]  # 특성 4개
y = df['species']                                 # 정답 (Adelie/Chinstrap/Gentoo, 글자 그대로 OK)
```

```python
# ------------------------------------------------------------
# [목적] 8:2 분리 -> 학습 -> 종별 성적표
# ------------------------------------------------------------
X_tr, X_te, y_tr, y_te = train_test_split(
    X, y, test_size=0.2, random_state=0, stratify=y)   # 종 비율 유지 8:2
model = RandomForestClassifier(random_state=0).fit(X_tr, y_tr)  # 학습
print(classification_report(y_te, model.predict(X_te)))        # 종별 정밀도·재현율·f1
```
