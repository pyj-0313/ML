```python
# ############################################################
# Level 2 — 와인 3종 분류 + 스케일링 + 클래스별 성적표
# ############################################################
# ------------------------------------------------------------
# [목적] 도구 불러오기 — 데이터 / 분리 / 전처리 / 모델 / 평가
# ------------------------------------------------------------
from sklearn.datasets import load_wine            # 와인 데이터 (3종)
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler  # 스케일링 (로지스틱은 크기에 민감)
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report # 클래스별 상세 성적표
```

```python
# ------------------------------------------------------------
# [목적] 데이터 준비 + 8:2 분리 + 스케일링
# ------------------------------------------------------------
X, y = load_wine(return_X_y=True)                 # 와인 성분 13개, 정답 3종
X_tr, X_te, y_tr, y_te = train_test_split(
    X, y, test_size=0.2, random_state=0, stratify=y)   # 종 비율 유지 8:2
sc = StandardScaler()
X_tr_s = sc.fit_transform(X_tr)                   # train만 기준 학습 + 변환
X_te_s = sc.transform(X_te)                       # test는 변환만
```

```python
# ------------------------------------------------------------
# [목적] 학습 후 성적표 (macro/weighted 평균 함께 확인)
# ------------------------------------------------------------
model = LogisticRegression(max_iter=5000).fit(X_tr_s, y_tr)   # 스케일한 값으로 학습
print(classification_report(y_te, model.predict(X_te_s)))     # 종별 + 평균 성적
```
