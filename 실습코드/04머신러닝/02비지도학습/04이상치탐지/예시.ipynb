```python
# ############################################################
# 이상치탐지 예시 — IsolationForest로 '튀는 값' 찾기 (-1/1)
# ############################################################
# ------------------------------------------------------------
# [목적] 도구 + 데이터 준비 (유방암 데이터로 유별난 샘플 찾기)
# ------------------------------------------------------------
from sklearn.datasets import load_breast_cancer   # 예시용 실제 데이터
from sklearn.preprocessing import StandardScaler  # 스케일링 (거리·밀도 기반이라 필요)
from sklearn.ensemble import IsolationForest      # 이상치 탐지 (빨리 고립되면 이상치)
X = load_breast_cancer().data                     # 특성 30개
```

```python
# ------------------------------------------------------------
# [목적] 스케일링 후 학습+판정 한 번에 (contamination = 이상치 비율 힌트)
# ------------------------------------------------------------
X_s = StandardScaler().fit_transform(X)           # 같은 잣대로 맞추기
iso = IsolationForest(contamination=0.05, random_state=0)  # '이상치 약 5%'라고 귀띔
pred = iso.fit_predict(X_s)                        # 학습+판정 -> 각 샘플에 -1 또는 1
```

```python
# ------------------------------------------------------------
# [목적] 결과 읽기 — -1(이상)/1(정상), 정상만 남겨 '데이터 정제'
# ------------------------------------------------------------
print('전체:', len(X_s), '| 이상치(-1):', (pred == -1).sum(), '| 정상(1):', (pred == 1).sum())
X_normal = X_s[pred == 1]                          # 정상만 남기기 (튀는 값 제거)
print('정제 후 남은 개수:', len(X_normal))
```

```python
# ============================================================
# [결과 해석]
#  · 569개 중 29개(약 5%)를 이상치(-1)로, 나머지 540개는 정상(1)으로 판정
#  · contamination=0.05로 '5%쯤 이상'이라 알려줬더니 딱 그만큼을 골라냄
#  · pred==1 만 남기면 '튀는 값 제거' -> 다음 분석을 위한 데이터 정제 완료
#  · 주의: 결과는 항상 -1(이상) / 1(정상) 두 종류뿐 (헷갈리지 말 것!)
# ============================================================
```
