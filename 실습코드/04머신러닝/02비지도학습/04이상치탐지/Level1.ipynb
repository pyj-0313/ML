```python
# ############################################################
# Level 1 — 정상 무리 + 이상치를 직접 만들어 탐지
# ############################################################
# ------------------------------------------------------------
# [목적] 도구 불러오기
# ------------------------------------------------------------
import numpy as np                                # 난수로 데이터 만들기
from sklearn.ensemble import IsolationForest      # 이상치 탐지
```

```python
# ------------------------------------------------------------
# [목적] 데이터 만들기 — 정상 100개(가운데) + 이상치 8개(넓게)
# ------------------------------------------------------------
rng = np.random.default_rng(0)                    # 난수 고정
normal   = rng.normal(0, 1, size=(100, 2))        # 정상 100개 (0 근처에 몰림)
outliers = rng.uniform(-6, 6, size=(8, 2))        # 이상치 8개 (-6~6으로 넓게)
X = np.vstack([normal, outliers])                 # 위아래로 쌓아 108개
```

```python
# ------------------------------------------------------------
# [목적] 학습+판정 -> 이상치 개수 세기
# ------------------------------------------------------------
iso = IsolationForest(contamination=0.1, random_state=0)  # 이상치 약 10% 가정
pred = iso.fit_predict(X)                          # -1=이상, 1=정상
print('이상치로 탐지된 개수:', (pred == -1).sum())
```

```python
# ============================================================
# [결과 해석]
#  · 108개 중 약 11개를 이상치로 탐지 (contamination 0.1 = 약 10%)
#  · 일부러 심어둔 이상치 8개 부근을 대체로 잡아냄
#  · 실전에선 어느 게 이상치인지 모른 채 찾는다 (여기선 이해를 위해 심어둠)
# ============================================================
```
