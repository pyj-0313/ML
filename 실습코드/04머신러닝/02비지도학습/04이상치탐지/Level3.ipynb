```python
# ############################################################
# Level 3 — 도매 고객 중 유별난 소비 고객 찾기 (공개 데이터)
# ############################################################
# ------------------------------------------------------------
# [목적] 도구 불러오기 — 표 / 전처리 / 모델
# ------------------------------------------------------------
import pandas as pd                               # 표(CSV) 다루기
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
```

```python
# ------------------------------------------------------------
# [목적] 데이터 불러오기 + 지출 특성 선택 + 스케일링
# ------------------------------------------------------------
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00292/Wholesale%20customers%20data.csv'
df = pd.read_csv(url)                              # 도매 고객 440명 (주소의 %20 = 빈칸)
feat = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']  # 품목별 연간 지출
X_s = StandardScaler().fit_transform(df[feat])    # 지출 단위가 커서 스케일링
```

```python
# ------------------------------------------------------------
# [목적] 판정 -> 표에 결과 열 붙이고 -> 이상 고객만 보기
# ------------------------------------------------------------
iso = IsolationForest(contamination=0.05, random_state=0)
df['anomaly'] = iso.fit_predict(X_s)              # 각 고객에 -1(이상)/1(정상)
print('이상치 개수:', (df['anomaly'] == -1).sum(), '/', len(df))
df[df['anomaly'] == -1][feat].head()              # 유별나게 많이/적게 쓴 고객
```
