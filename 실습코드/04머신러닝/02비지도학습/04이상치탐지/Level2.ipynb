```python
# ############################################################
# Level 2 — 와인 데이터 + '가장 이상한 것' 콕 집기 (점수)
# ############################################################
# ------------------------------------------------------------
# [목적] 도구 불러오기 — 배열 / 데이터 / 전처리 / 모델
# ------------------------------------------------------------
import numpy as np
from sklearn.datasets import load_wine            # 와인 성분 13개
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
```

```python
# ------------------------------------------------------------
# [목적] 스케일링 후 판정 + '이상함의 정도(점수)'까지 받기
# ------------------------------------------------------------
X = load_wine().data                              # 특성 13개
X_s = StandardScaler().fit_transform(X)           # 단위 제각각이라 스케일링
iso = IsolationForest(contamination=0.05, random_state=0)
pred = iso.fit_predict(X_s)                        # -1=이상, 1=정상
score = iso.decision_function(X_s)                # 점수 낮을수록 더 이상함

print('이상치 개수:', (pred == -1).sum(), '/', len(X_s))
print('가장 이상한 5개 인덱스:', np.argsort(score)[:5])   # 점수 낮은 순 5개
```

```python
# ============================================================
# [결과 해석]
#  · 178개 중 9개를 이상치로 판정 (약 5%)
#  · decision_function 점수가 낮을수록 '더 유별남' -> argsort로 가장 이상한 5개를 콕 집음
#  · -1/1 판정만이 아니라 '순위'까지 알 수 있어, 우선 살펴볼 대상을 고르기 좋음
# ============================================================
```
